## Load the image and packages



```python
import os
import sys 
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
sys.path.append(os.path.join('..', 'packages'))
sys.path.append(os.path.join('..', 'data', 'input'))
import tekenfuncties_2 as tf

#read image into np array 
loc_input_img = os.path.join('..', 'data', 'input', 'rabbit2.jpg')
image = Image.open(loc_input_img)
np_image = np.array(image)
```

## Show the normal image

```python
tf.show_image(np_image)
```

## Show coloured image (use "R", "G", "B")

```python
# show red image 
np_image_K = tf.verkleur(np_image, "R")                    
tf.show_image(np_image_K)
```

## Grid with flips (3 different grids)

```python
#grid_with_flips1
haas_matrix =  [[0 for i in range(7)] for j in range(3)] 
grid_image = tf.grid_with_flips(np_image, haas_matrix)
tf.show_image(grid_image)

#grid_with_flips2
haas_matrix_2 = [[j %2 for i in range(7)] for j in range(4)]
grid_image = tf.grid_with_flips(np_image, haas_matrix_2)
tf.show_image(grid_image)

#grid_with_flips3
haas_matrix_3 =  [[(i+j) % 4 for i in range(10)] for j in range(10)] 
grid_image = tf.grid_with_flips(np_image, haas_matrix_3)
tf.show_image(grid_image)
```

## show colourful big one, 4x4 and 5x5

```python
haas_kleuren = ['b', 'b', 'b', 'b', 'r', 'r', 'g', 'g', 'g', 'g', 'r', 'r']
grid_image = tf.create_colorful_big_one(np_image, haas_kleuren)
tf.show_image(grid_image)

haas_kleuren_2 = ['b','b', 'b', 'b', 'b', 'r','r', 'r', 'g','g', 'g', 'g', 'g', 'r', 'r', 'r']
grid_image = tf.create_colorful_big_one(np_image, haas_kleuren_2)
tf.show_image(grid_image)
```

## zoom

```python
zoem_image = tf.zoom_image(np_image, (80,65), 4)    
tf.show_image(zoem_image)

zoem_image = tf.zoom_image(np_image, (25,25), 2)    
tf.show_image(zoem_image)
```

## Rotate

```python
rotated_image = tf.rotate_image(np_image, 45)
tf.show_image(rotated_image)

rotated_image = tf.rotate_image(np_image, -20)
tf.show_image(rotated_image)

```
